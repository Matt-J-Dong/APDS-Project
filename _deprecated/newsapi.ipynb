```python
import requests

API_KEY = '93f950e09bdf4c5fb8ec6aecd61ee6fb'
BASE_URL = 'https://newsapi.org/v2'

def get_stock_news(query, page_size=100):
    """
    Fetch recent news articles matching a given query (e.g., 'AAPL' or 'Apple Inc').
    """
    endpoint = f"{BASE_URL}/everything"
    params = {
        'q': query,
        'apiKey': API_KEY,
        'pageSize': page_size,
        'sortBy': 'publishedAt',
        'language': 'en'
    }
    response = requests.get(endpoint, params=params)
    return response.json()

news_response = get_stock_news("AAPL")
articles = news_response.get('articles', [])
for article in articles:
    print(article['title'], "-", article['source']['name'])
```

```python
import requests
import csv

API_KEY = '93f950e09bdf4c5fb8ec6aecd61ee6fb'
BASE_URL = 'https://newsapi.org/v2'

def get_stock_news(query, page_size=100, page=1):
    """
    Fetch recent news articles matching a given query
    (only one page of up to 100 articles).
    """
    endpoint = f"{BASE_URL}/everything"
    params = {
        'q': query,
        'apiKey': API_KEY,
        'pageSize': page_size,  # 100 max
        'page': page,           # only the first page
        'sortBy': 'publishedAt',
        'language': 'en'
    }
    response = requests.get(endpoint, params=params)
    return response.json()

# Fetch only one page of up to 100 articles
news_response = get_stock_news("AAPL", page_size=100, page=1)
articles = news_response.get('articles', [])

# Write to CSV
with open('newsapi_AAPL.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Time Data', 'Headline']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for article in articles:
        published_at = article.get('publishedAt', '')
        # Merge title + description into one field
        title = article.get('title', '')
        description = article.get('description', '')
        merged_headline = f"{title}. {description}"
        
        writer.writerow({
            'Time Data': published_at,
            'Headline': merged_headline
        })
```
